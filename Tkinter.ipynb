```python
import tkinter as tk
from tkinter import *

from PIL import Image, ImageTk

def resize_image(file, x, y):
    
    from PIL import Image
    
    return Image.open(file).resize((x, y))
    
    
def exit_app():
    valor = messagebox.askquestion("Salir. Â¿Seguro?")
    if valor == "yes":
        window.destroy()
    
def menu_bar(window):
    menu_bar = tk.Menu(window)
    window.config(menu = menu_bar,  width = 450, height = 250)
    
    menu_inicio = tk.Menu(menu_bar)
    menu_bar.add_cascade(label = 'Inicio', menu = menu_inicio)
    
    menu_inicio.add_command(label = 'Create Regiester in BBDD')
    menu_inicio.add_command(label = 'Delete Regiester in BBDD')
    menu_inicio.add_command(label = 'Exit')
    
```

```python
window = tk.Tk()
window.title('Robert_Johnson')
window.iconbitmap('images/demon.ico')
window.geometry("500x300")
window.config(bg = 'black', cursor = "cross")
window.resizable(1,1)

# -------------------------- FRAME ----------------------------------

frame = tk.Frame(window)
frame.config(width = 450, height = 250, bg = 'yellow')
frame.pack(expand = 1)
frame.config(bd = '20', relief='sunken', cursor='heart')

# ------------------------ LABEL -----------------------

imagen_label = PhotoImage(file = 'images/tri.gif')
label = Label(frame, text = 'Joseph Tartini')
label.image = imagen_label
#label.pack()
label.config(fg="red", bg="white", font=("Arial",32))

# --------------------- TEXTBOX ------------------------

textbox_1 = Entry(frame)
textbox_1.grid(row = 0, column = 1)

textbox_password = Entry(frame)
textbox_password.grid(row = 1, column = 1)
textbox_password.config(show = "*")

textbox_2 = Entry(frame)
textbox_2.grid(row = 2, column = 1)

textbox_3 = Entry(frame)
textbox_3.grid(row = 3, column = 1)

big_textbox = Text(frame, width = 20, height = 5)
big_textbox.grid(row = 4, column = 1)

# ------------- LABELS TEXTBOX ---------------

box_1_label = Label(frame, text = "Name: ")
box_1_label.config(fg = "red", bg = "black", font = ("Arial", 12))
box_1_label.grid(row = 0, column = 0, sticky = 'e')

box_password = Label(frame, text = "Password: ")
box_password.grid(row = 1, column = 0, sticky = 'e')


box_2_label = Label(frame, text = "Last_Name: ")
box_2_label.grid(row = 2, column = 0, sticky = 'e')

box_3_label = Label(frame, text = "Age: ")
box_3_label.grid(row = 3, column = 0, sticky = 'e')

big_textbox_label = Label(frame, text = "Comment: ")
big_textbox_label.grid(row = 4, column = 0)


# ------------------- BUTTON -----------------------


icon_button = ImageTk.PhotoImage(resize_image('images/eye.png', 50, 50))
button = Button(frame, text="Send", command=window.destroy, image = icon_button)

button.config(bg = 'green', fg = 'red', font = ("Arial", 12))
button.grid(row = 5, column = 3)


# ---------------- SCROLL BAR -------------------
    
    
scroll_v = Scrollbar(frame, command = big_textbox.yview)
scroll_v.grid(row = 4, column = 2, sticky = "ns")
big_textbox.config(yscrollcommand = scroll_v.set)

# ---------------- PRUEBAS ------------------

#PRUEBA = ttk.Treeview(frame)
#PRUEBA.grid(row = 6, column = 3)



window.mainloop()

```
